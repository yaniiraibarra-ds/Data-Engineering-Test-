# **Ejercicio 4: Web Scraping con Requests**

1. Realizar una solicitud GET a la URL https://coinmarketcap.com/currencies/bitcoin/ utilizando la biblioteca requests.
2. Analizar el contenido de la página web para extraer el precio actual del Bitcoin.
3. Imprimir el precio en la consola.

Importar librerías

```python
import requests
from bs4 import BeautifulSoup
```

Obtener y mostrar el precio actual de Bitcoin desde CoinMarketCap mediante web scraping

```python

#URL de la página de CoinMarketCap para Bitcoin
url = 'https://coinmarketcap.com/currencies/bitcoin/'

# Realiza una solicitud GET
headers = {'User-Agent': 'Mozilla/5.0'}  # Evita bloqueos
response = requests.get(url, headers=headers)

# Verificar solicitud exitosa
if response.status_code == 200:
# Analizar el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

# Nuevo selector de la pagina, buscar el span con data-test="text-cdp-price-display"
    price_tag = soup.find('span', attrs={'data-test': 'text-cdp-price-display'})
    if price_tag:
        precio_bitcoin = price_tag.text.strip()
        print(f"Precio actual de Bitcoin (BTC): {precio_bitcoin}")
    else:
        print("No se pudo encontrar el precio en la página. La estructura HTML pudo haber cambiado.")
else:
    print(f'Error al hacer la solicitud. Código de respuesta: {response.status_code}')
```
